```python
import os
import pickle
```

```python
datasetwise = {}
full_saver=[]
for data in['CBF','CharacterTrajectories','Coffee','ECG5000','ElectricDevices','FordA','GunPoint','NATOPS', 'UWaveGestureLibrary']: #os.listdir('./Results/mutate_both/'):
    #print(data)
    if os.path.isdir(f'./Results/mutate_both/{data}'):
        dw_saver=[]
        for item in os.listdir(f'./Results/mutate_both/{data}'):
            if item.startswith('Time'):
                d = pickle.load(open(f'./Results/mutate_both/{data}/{item}','rb'))
                full_saver.append(d.total_seconds())
                dw_saver.append(d.total_seconds())
        datasetwise[data]=dw_saver
```

```python
import numpy as np 
np.mean(full_saver)/60
#np.std(full_saver)
```
