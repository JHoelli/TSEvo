```python
import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches
import numpy as np
verts = [ (x, y) for x,y in zip (train_x[100][0],train_x[100][1]) ]
code = [Path.MOVETO]
code.extend(np.repeat(Path.LINETO,len(verts)-1 ))
p=Path(verts,code)
fig, ax = plt.subplots()
patch = patches.PathPatch(p)#, facecolor='orange', lw=2)
ax.add_patch(patch)
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
plt.show()
```

```python
# plot control points and connecting lines
%matplotlib inline
x, y = zip(*p.vertices)
plt.plot(x, y, 'go-')
plt.show()
```

```python
x_calc=[0]
counter=0
for x in train_x[100][0]:
    x_calc.append(x+x_calc[counter])
    counter+=counter
y_calc=[0]
counter=0
for y in train_x[100][1]:
    y_calc.append(y+y_calc[counter])
    counter+=counter
```

```python
# plot control points and connecting lines
%matplotlib inline
plt.plot(x_calc, y_calc, 'go-')
plt.show()
```

```python
basis = (0,0)
verts=[basis]
counter=0
for x,y in zip(train_x[100][0],train_x[100][1]):
    verts.append((verts[counter][0]+x,verts[counter][0]+y))
    counter+=counter
```

```python
x_new=[a[0] for a in verts]
y_new=[a[1] for a in verts]
```
